```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10, 8]
```

```python
df = pd.read_excel(r'all_results.xlsx')
```

```python
list(set(df.mtry))
```

```python
m_list = list(set(df.mtry))
m_list.sort()

colors = ['#fde725', '#b5de2b', '#1E8251', '#35b779', '#31688e', '#482878']

for m, col in zip(m_list, colors):
    tmp = df[df.mtry == m].reset_index(drop = True)
    plt.scatter(tmp.n, tmp.AUC, marker = 'o', c = col, s = plt.rcParams['lines.markersize'] ** 3, label = f'mtry={int(m)}', 
               edgecolors='k', linewidths=.5)
plt.scatter([500], [0.8972], marker = '*', c = '#ed1000', s = plt.rcParams['lines.markersize'] ** 3.35, label = f'default settings', 
            edgecolors='k', linewidths=.5)
plt.legend()
plt.xlabel('n', fontsize=12)
plt.ylabel('AUC', fontsize=12)
plt.title('Bayesian optimisation results', fontsize=15)
plt.show()
```
